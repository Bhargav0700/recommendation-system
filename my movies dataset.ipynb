```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
movies=pd.read_csv("my_movies.csv")
```

```python
movies
```

```python
movies.describe()
```

```python
movies=movies.iloc[:,5:]
movies
```

```python
movies.info()
```

```python
movies.dtypes
```

```python
confidence=[0.1,0.3,0.5,0.6,0.7,0.8,0.9,1.0]
ap=[]
for i in confidence:
    ap=apriori(movies,min_support=0.5,use_colnames=True)
    rule1= association_rules(ap,min_threshold=i)
    ap.append(rule1)
    
ap
```

```python
ap.sort_values("support")
```

```python
movies
```

```python
frequent_itemsets=apriori(movies,min_support=0.2,use_colnames=True)
frequent_itemsets
```

# confidence

```python
rule=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
rule
```

```python
rule.sort_values("confidence")
```

# LIFT

```python
rule2=association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rule2
```

```python
rule2.sort_values("lift")
```

  2.lift must be greater than one and upto inf
